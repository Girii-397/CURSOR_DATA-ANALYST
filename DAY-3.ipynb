```python
data = pd.read_csv('/content/airqualityindexdata/AQI Data Set.csv')
```

```python
data.head()
```

```python
import pandas as pd
import numpy as np
from scipy import stats

# Calculate mean
mean = data['AQI'].mean()

# Calculate median
median = data['AQI'].median()

# Calculate mode
mode = data['AQI'].mode()[0]

# Calculate variance
variance = data['AQI'].var()

# Calculate standard deviation
std_dev = data['AQI'].std()

# Calculate range
data_range = data['AQI'].max() - data['AQI'].min()

# Calculate skewness
skewness = data['AQI'].skew()

# Detect outliers using IQR method
Q1 = data['AQI'].quantile(0.25)
Q3 = data['AQI'].quantile(0.75)
IQR = Q3 - Q1
outliers = data[(data['AQI'] < (Q1 - 1.5 * IQR)) | (data['AQI'] > (Q3 + 1.5 * IQR))]

# Print the results
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
print("Range:", data_range)
print("Skewness:", skewness)
print("Number of Outliers:", len(outliers))
```
